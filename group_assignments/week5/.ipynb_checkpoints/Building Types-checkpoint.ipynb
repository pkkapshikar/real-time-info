```python
import osmnx as ox

import pandas as pd

import plotly.express as px

import geopandas as gpd

import contextily as ctx

import matplotlib.pyplot as plt

import cartopy 
```

```python
address = 'Oakland, CA'
```

```python
%%time
# %%time is a magic command to see how long it takes this cell to run 

# get the data from OSM that are tagged as 'building' for a 1000m X 1000m square area
osm = ox.geometries_from_address(address,tags={'building':True},dist=1000)
```

```python
# how many rows and columns?
osm.shape
```

```python
# what is the datatype?
type(osm)
```

```python
# show me 10 random rows
osm.sample(10)
```

```python
osm.info(verbose=True, show_counts=True)
```

```python
## subset it
columns_to_keep = ['geometry','building','height']
osm = osm[columns_to_keep]
osm.head(10)
```

```python
osm.loc[osm['building']=='yes','building'] = 'unspecified'
```

```python
osm_building_counts1 = osm.value_counts(['building']).reset_index(name="count")
osm_building_counts1
```

```python
osm_building_counts = osm.value_counts(['building']).reset_index(name="count")
buildings = ['apartments','residential','house','commercial','industrial','supermarket','school']
osm_building_counts = osm_building_counts.loc[osm_building_counts['building'].isin(buildings)] 
osm_building_counts
```

```python
osm_filtered = osm.loc[osm['building'].isin(buildings)]
osm_filtered
```
