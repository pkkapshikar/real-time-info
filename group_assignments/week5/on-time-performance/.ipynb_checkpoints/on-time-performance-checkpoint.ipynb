```python
import osmnx as ox

import pandas as pd

import plotly.express as px

import geopandas as gpd

import contextily as ctx

import matplotlib.pyplot as plt

import fiona

import seaborn as sns

import numpy as np
```

```python
df_19 = pd.read_csv('On Time Performance_Line 19 (2022).csv')
df_20 = pd.read_csv('On Time Performance_Line 20 (2022).csv')
df_51A = pd.read_csv('On Time Performance_Line 51A (2022).csv')
```

```python
df_19 = df_19.dropna()
df_20 = df_20.dropna()
df_51A = df_51A.dropna()
```

```python
df_19['Eastbound OTP'] = df_19['Eastbound OTP']*100
df_19['Westbound OTP'] = df_19['Westbound OTP']*100
df_19.head(12)
```

```python
ax1 = df_19.plot.line(x='Month',ylabel='Percent',rot=90)
ax1.set_ylim(ymin=0, ymax=100)
ax1.set_xticks(np.linspace(0,11,12), labels=df_19['Month'])
ax1.set_title('AC Transit 19 On-Time Performance')
ax1.legend(loc = 'lower right')
```

```python
df_20['Eastbound OTP'] = df_20['Eastbound OTP']*100
df_20['Westbound OTP'] = df_20['Westbound OTP']*100
df_20.head(12)
```

```python
ax2 = df_20.plot.line(x='Month',ylabel='Percent',rot=90)
ax2.set_ylim(ymin=0, ymax=100)
ax2.set_xticks(np.linspace(0,11,12), labels=df_20['Month'])
ax2.set_title('AC Transit 20 On-Time Performance')
ax2.legend(loc = 'lower right')
```

```python
df_51A['Eastbound OTP'] = df_51A['Eastbound OTP']*100
df_51A['Westbound OTP'] = df_51A['Westbound OTP']*100
df_51A.head(12)
```

```python
ax3 = df_51A.plot.line(x='Month',ylabel='Percent',rot=90)
ax3.set_ylim(ymin=0, ymax=100)
ax3.set_xticks(np.linspace(0,11,12), labels=df_51A['Month'])
ax3.set_title('AC Transit 51A On-Time Performance')
ax3.legend(loc = 'lower right')
```
